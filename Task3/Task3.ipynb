```python
# Importing the required libraries
import pandas as pd
```

```python
# load datasets
rating = pd.read_csv("ratings.csv")
rating.head(2)
```

```python
print("The Shape is:")
rating.shape
```

```python
movies = pd.read_csv("movies.csv")
movies = movies.iloc[:, 0:2]
movies.head(2)
```

```python
print("The Shape is:")
movies.shape
```

```python
# Merge
df = pd.merge(rating, movies, on='movieId')
df.head(10)
```

```python
print("The Shape is:")
df.shape
```

```python
# We won't use the timestamp, let's drop it
df.drop(["timestamp"], axis=1, inplace=True)
df.head(7)
```

```python
# Set Up Our Recommendation System
moviemat = df.pivot_table(index='userId', columns='title', values='rating')
moviemat.head(4)
```

```python
print("The type is:")
type(moviemat)
```

```python
# Let's look at the user ratings of the film Braveheart (1995):
Braveheart_user_ratings = moviemat["Braveheart (1995)"]
Braveheart_user_ratings.head()
```

```python
# Let's calculate the correlation of all films
# with the Braveheart film using the corrwith() method:
similar_to_Braveheart = moviemat.corrwith(Braveheart_user_ratings)
similar_to_Braveheart.head(30)
#It gives the following error due to NaN values.
```


According to this result, the correlation between the Braveheart (1995) film  and the film
"burbs, The (1989)" is 0.660184. In other words, 66% of the people who watched
the Braveheart (1995) film also watched the film "burbs, The (1989)".

On the other hand, the correlation between the film "Braveheart (1995)" and the film "102 Dalmatians (2000)"
was -0.970725. In other words, 97% of the people who watched the Braveheart  film did not watch
 the film "102 Dalmatians (2000)".

```python
print("The type is:")
type(similar_to_Braveheart)
```

```python
# The list we've got is a Pandas series. If we convert it to df, we can work more easily.
corr_Braveheart = pd.DataFrame(similar_to_Braveheart, columns=['Correlation'])
corr_Braveheart.dropna(inplace=True)
corr_Braveheart.head()
```

```python
# Let's list the data frame we obtained and see which film it suggests as the closest to Braveheart
corr_Braveheart.sort_values('Correlation', ascending=False).head(10)
```

```python
# Let's find the average (mean value) rating of each film
ratings = pd.DataFrame(df.groupby('title')['rating'].mean())
ratings
```

```python
# Let's sort it
ratings.sort_values('rating', ascending=False).head(20)
```

```python
ratings['rating_vote_number'] = pd.DataFrame(df.groupby('title')['rating'].count())
ratings.head()
```

```python
# Now let's take a look at the most voted films in descending order¶
ratings.sort_values('rating_vote_number', ascending=False).head()
```

```python
# add the rating_vote_number column to our corr_Braveheart dataframe (with join)
corr_Braveheart.sort_values('Correlation', ascending=False).head(10)
```

```python
corr_Braveheart = corr_Braveheart.join(ratings['rating_vote_number'])
corr_Braveheart.head()
```

NOTE: Films with a small number of votes can mislead us.
For example, the fact that 1-2 people gave a film 5 does not mean the same thing as the fact that
100 people gave it 5. For this reason, it would be much more accurate to decide by looking at
the average ratings of films with more than 100 votes

```python
final = corr_Braveheart[corr_Braveheart['rating_vote_number']> 100].sort_values('Correlation', ascending=False)
final.head()
```
